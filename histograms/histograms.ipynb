```python
import cv2
import numpy as np

#need matplotlib to plot
from matplotlib import pyplot as plt

image = cv2.imread('./images/arch.jpg')

#cv2.calcHist(image, mask, histSize, rangesp, hist[, accumulate]])
histogram = cv2.calcHist([image],[0], None,[256], [0, 256])

#we plot as histogram, ravel() flatens our image array
#ravel() takes, 2D array and makes it 1D
plt.hist(image.ravel(), 256, [0, 256])
plt.show()

```

```python
#viewing separate colors channels, using LOOP
color = ('b', 'g', 'r')

#we now separate the colors and plot each in the histogram
for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image],[i], None,[256], [0, 256])
    plt.plot(histogram2, color = col)
    plt.xlim([0, 256])
plt.show() 
```
